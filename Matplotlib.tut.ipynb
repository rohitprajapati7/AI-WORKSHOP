```python
x=[2,9,3,7]
y=[7,15,5,9]
plt.plot(x,y,label='railway track')
plt.scatter(x,y,s=200,label='railway station')#scatter use for points
#if you want increase a size of scatter points
plt.xlabel('Distance')
plt.ylabel('Time')
plt.grid(color='g')#draw a grid
plt.legend() #it will show label in the plot
plt.show()
```

```python
x=[2,9,3,7]
y=[7,15,5,9]
plt.plot(x,y,label='railway track')
plt.scatter(x,y,s=200,label='railway station',marker='*',color='r')#scatter use for points 
#marker is keyword it change points to any onther symbol
#if you want increase a size of scatter points use s(size) and give value
plt.xlabel('Distance')
plt.ylabel('Time')
plt.grid(color='g')#draw a grid
plt.legend() #it will show label in the plot
plt.show()
```

```python
#BAR CHART 
```

```python
x=[2,9,3,7]
y=[7,15,5,9]
plt.plot(x,y,label='railway track')
plt.scatter(x,y,s=200,label='railway station',marker='*',color='r')#scatter use for points 
#marker is keyword it change points to any onther symbol
#if you want increase a size of scatter points use s(size) and give value
plt.bar(x,y,label='height of railway station') #to draw bar chart
plt.xlabel('Distance')
plt.ylabel('Time')
plt.grid(color='g')#draw a grid
plt.legend() #it will show label in the plot
plt.show()
```

```python
#this will be another example of cricket data visualization
players=['dhoni','virat','rohit','yuvraj']
scores=[90,80,100,90]
#now plotting it
plt.xlabel('players')
plt.ylabel('scores')
plt.bar(players,scores,label='2020 scores of indian cricket team')
# plt.grid(color='g')
plt.legend()
plt.show()
```

```python
#pie chart

#this will be another example of cricket data visualization
players=['dhoni','virat','rohit','yuvraj']
scores=[90,80,100,90]
#now plotting it
plt.xlabel('players')
plt.ylabel('scores')
plt.pie(scores,labels=players)
# plt.grid(color='g')
plt.legend()
plt.show()
```

```python
#pie chart

#this will be another example of cricket data visualization
players=['dhoni','virat','rohit','yuvraj']
scores=[90,80,100,90]
#now plotting it
plt.xlabel('players')
plt.ylabel('scores')
plt.pie(scores,labels=players,explode=(0.2,0,0,0),shadow=True,autopct='%1.1ff%%')
# plt.grid(color='g')
plt.legend()
plt.show()
```
